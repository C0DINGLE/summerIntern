##### Lets create a column for total scores out of 100 for each student.

```python
data['final_scores'] = (data['math_score']+data['writing_score']+data['reading_score'])*100/300
# data.head()
```

now lets create a grading system for better comparison of marks from this point onwards.

for instance:
 * 81-100 :Grade A
 *  71-80 : Grade B
 *  61-70 : Grade C
 *  51-60 : Grade D
 *  41-50 : Grade E
 *  0 - 40  : Grade F

```python
def gradingSys(scores):
    if scores >= 81 and scores <= 100:
        return 'Grade A'
    elif scores >= 71 and scores < 81:
        return 'Grade B'
    elif scores >= 61 and scores < 71:
        return 'Grade C'
    elif scores >= 51 and scores < 61:
        return 'Grade D'
    elif scores >= 41 and scores < 51:
        return 'Grade E'
    elif scores >= 0 and scores < 41:
        return 'Grade F'
    
data['grades']=data['final_scores'].apply(gradingSys)
# data.head()
```

Lets see how the grades are distributed across the entire dataset.

```python
data['grades'].value_counts()
```

```python
data['grades'].value_counts().plot.pie(startangle = 20,shadow = True, autopct="%1.1f%%", explode = (0.2,0.05,0.1,0,0,0)).set_title('Grades',fontsize = 20)
plot.axis('equal')
plot.show()
```

Thus we can observe from this that **Grade C (Marking 60 - 70)** and **Grade B (Marking 70-80)** are most frequent across the data sets subjects.

Lets say that the Grade F is failing grade. Lets check the statistics of failing students.

```python
plot.rcParams['figure.figsize'] = (10,5)
sns.countplot(x="gender", hue="grades", data=data).set_title("Gender Vs Grades",fontsize = 40)
plot.show()
```

```python
data[(data.grades == 'Grade F')]['gender'].value_counts()
```

Thus **females are overall better performers** but when it comes to **failed students, females lead there too**.

```python
plot.rcParams['figure.figsize'] = (18,10)
sns.countplot(x="parental_level_of_education", hue="grades", data=data).set_title("Gender Vs Grades",fontsize = 40)
plot.show()
```

This clearly shows that students of parents holding **an undergraduate degree or higher** ***are significantly less in terms of failing grades***.
